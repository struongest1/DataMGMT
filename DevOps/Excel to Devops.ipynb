```python
import pandas as pd
import requests
from requests.auth import HTTPBasicAuth
import json

# CONFIGURATION
AZURE_ORG = 'ECCC-DevOps  # e.g. 'myorg'
AZURE_PROJECT = 'Data Management Development'   # e.g. 'MyProject'
PAT = 'AcDXc3gGA2YuUc1AdUosBuMiGWLqYeSDFnqmqr1QpSASNJJrO5VQJQQJ99BDACAAAAAaaZ9IAAASAZDO3JFZ'           # your Personal Access Token
EXCEL_FILE = 'tasks.xlsx'
WORK_ITEM_TYPE = 'Task'          # or 'Product Backlog Item', 'Bug', etc.

# Azure DevOps API URL
AZURE_URL = f'https://dev.azure.com/ECCC-DevOps/Data%20Management%20Development/_apis/wit/workitems/${{WORK_ITEM_TYPE}}?api-version=7.0'

# Read Excel file
df = pd.read_excel(EXCEL_FILE)

# Loop through each row and create a work item
for index, row in df.iterrows():
    title = row['Title']
    description = row.get('Description', '')
    assigned_to = row.get('AssignedTo', None)

    # Create JSON payload in patch format
    patch_data = [
        {
            "op": "add",
            "path": "/fields/System.Title",
            "from": None,
            "value": title
        },
        {
            "op": "add",
            "path": "/fields/System.Description",
            "from": None,
            "value": description
        }
    ]

    if assigned_to:
        patch_data.append({
            "op": "add",
            "path": "/fields/System.AssignedTo",
            "from": None,
            "value": assigned_to
        })

    # Make the API request
    response = requests.patch(
        AZURE_URL,
        auth=HTTPBasicAuth('', PAT),
        headers={
            'Content-Type': 'application/json-patch+json'
        },
        data=json.dumps(patch_data)
    )

    if response.status_code == 200 or response.status_code == 201:
        print(f"✅ Created: {title}")
    else:
        print(f"❌ Failed to create: {title} - {response.status_code}")
        print(response.json())
```
